```python
#methods
#get
#dictobj.get(key)
d={"name":["abc","def","pqr"],"courses":["da","ds","de"] ,"fees" :[23000,34000,30000]}
print(d.get("name"))
for i in d.get("name"):
    print(i)
   
```

```python
#keys
#dictobj.keys()
d={"name":["abc","def","pqr"],"courses":["da","ds","de"] ,"fees" :[23000,34000,30000]}
print(d.keys())
for i in d.keys():
    print(i)
```

```python
#values
#dictobj.values()
d={"name":["abc","def","pqr"],"courses":["da","ds","de"] ,"fees" :[23000,34000,30000]}
print(d.values())
for i in d.values():
    print(i)
```

```python
#items
#dictobj.items()
d={"name":["abc","def","pqr"],"courses":["da","ds","de"] ,"fees" :[23000,34000,30000]}
print(d.items())
for i in d.items():
    print(i)
```
