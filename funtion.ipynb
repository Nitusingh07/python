```python
def leap_year(year):
    if year%4==0:
        if year%100==0:
            if year%400==0:
                return True 
            else:
                return False
        else:
            return True
    else:
        return False
        
```

```python
num1=int(input("enter any number:"))
if leap_year(num1):
    print(num1,"leap year")
else:
    print(num1," not leap year")
```

```python
def arm(num):
    digits = len(str(num))
    total = 0
    temp = num

    while temp > 0:
        digit = temp % 10
        total += digit ** digits
        temp //= 10

    return total == num


num = int(input("Enter a number: "))

if arm(num):
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")

    
```

```python
def feb(n):
    if n==1:
        return [0]
    elif n==2:
        return [0,1]
    febo=[0,1]
    for i in range(2,n):
        next_term=febo[-1]+febo[-2]
        febo.append(next_term)
    
    return febo
    
    
    
```
