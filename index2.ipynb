```python
#slicing - extract substring from given string
#str obj[start:stop:step]
#start-0
#stop - end of the string--- given stop condition- n-1
#step- +1

s="welcome to bhopal"
print(s[::])
print(s[::-1]) #start -1 step -1 stop -end of string
print(s[10::2])
```

```python
#indirect
s="decode"
for i in range(len(s)):
    print(i,s[i])
```
