```python
s="welcome to the city of lakes"
vowel=0
space=0
consonant=0
for i in s:
    if i in "aeiou":
        vowel=vowel+1
    elif i==" ":
        space=space+1
    else:
        consonant=consonant+1
print(vowel)
print(space)
print(consonant)
```
